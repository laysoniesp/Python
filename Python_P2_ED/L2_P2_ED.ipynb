```python
import math

class Circulo:
    def __init__(self, raio):
        self.raio = raio

    def calcular_area(self):
        return math.pi * (self.raio ** 2)

raio_input = float(input("Digite o valor do raio: "))
valor_circulo = Circulo(raio_input)
print(f"A área do círculo é: {valor_circulo.calcular_area():.2f}")
```

```python
class Livro:
    def __init__(self, titulo, autor):
        self.titulo = titulo
        self.autor = autor

    def detalhes(self):
        return f"Título: {self.titulo}, Autor: {self.autor}"

titulo_input = input("Digite o título do livro: ")
autor_input = input("Digite o autor do livro: ")

meu_livro = Livro(titulo_input, autor_input)

print(meu_livro.detalhes())
```

```python
class Retangulo:
    def __init__(self, base, altura):
        self.base = base
        self.altura = altura

    def calcular_area(self):
        return self.base * self.altura

base = float(input("Digite a base do retângulo: "))
altura = float(input("Digite a altura do retângulo: "))
print(f"A área do retângulo é: {Retangulo(base, altura).calcular_area()}")
```

```python
class ContaBancaria:
    def __init__(self, titular, saldo=0):
        self.titular = titular
        self.saldo = saldo

    def depositar(self, valor):
        self.saldo += valor
        print(f"Depósito de R$ {valor:.2f} realizado com sucesso.")

    def sacar(self, valor):
        if valor <= self.saldo:
            self.saldo -= valor
            print(f"Saque de R$ {valor:.2f} realizado com sucesso.")
        else:
            print(f"Saldo insuficiente. Saldo atual: R$ {self.saldo:.2f}")

def converter_para_float(texto):
    texto_limpo = texto.replace('.', '').replace(',', '.')
    return float(texto_limpo)

titular_input = input("Digite o nome do titular da conta: ")
saldo_raw = input("Digite o saldo inicial (ex: 1.000,00): ")
saldo_input = converter_para_float(saldo_raw)

minha_conta = ContaBancaria(titular_input, saldo_input)

opcao = input("Você deseja (S)acar, (D)epositar ou (N)enhum? ").upper()

if opcao == 'D':
    valor_deposito = converter_para_float(input("Digite o valor do depósito: "))
    minha_conta.depositar(valor_deposito)
elif opcao == 'S':
    valor_saque = converter_para_float(input("Digite o valor do saque: "))
    minha_conta.sacar(valor_saque)

print(f"Saldo final de {minha_conta.titular}: R$ {minha_conta.saldo:.2f}")
```

```python
class Pessoa:
    def __init__(self, nome, idade):
        self.nome = nome
        self.idade = idade

    def falar(self):
        print(f"Olá, meu nome é {self.nome} e eu tenho {self.idade} anos.")

nome_input = input("Digite o nome: ")
idade_input = input("Digite a idade: ")

pessoa = Pessoa(nome_input, idade_input)
pessoa.falar()
```

```python
class Produto:
    def __init__(self, nome, preco, quantidade):
        self.nome = nome
        self.preco = preco
        self.quantidade = quantidade

    def calcular_total(self):
        return self.preco * self.quantidade

nome_prod = input("Produto: ")
p_preco_raw = input("Preço: ").replace(',', '.')
p_preco = float(p_preco_raw)

p_qtd = int(input("Quantidade: "))
prod = Produto(nome_prod, p_preco, p_qtd)
print(f"Total em estoque: R$ {prod.calcular_total():.2f}")
```

```python
class Aluno:
    def __init__(self, nome, notas):
        self.nome = nome
        self.notas = notas

    def calcular_media(self):
        return sum(self.notas) / len(self.notas)

nome_aluno = input("Nome do aluno: ")
entrada_notas = input("Digite as notas separadas por espaço (ou vírgula): ")

notas_limpas = entrada_notas.replace(',', ' ')
lista_notas = [float(nota) for nota in notas_limpas.split()]

aluno = Aluno(nome_aluno, lista_notas)
print(f"Média de {aluno.nome}: {aluno.calcular_media():.1f}")
```

```python
class Triangulo:
    def __init__(self, lado1, lado2, lado3):
        self.lado1 = lado1
        self.lado2 = lado2
        self.lado3 = lado3

    def calcular_perimetro(self):
        return self.lado1 + self.lado2 + self.lado3

l1 = float(input("Lado 1: "))
l2 = float(input("Lado 2: "))
l3 = float(input("Lado 3: "))
tri = Triangulo(l1, l2, l3)
print(f"Perímetro: {tri.calcular_perimetro()}")
```

```python
class Funcionario:
    def __init__(self, nome, salario, cargo):
        self.nome = nome
        self.salario = salario
        self.cargo = cargo

    def aumentar_salario(self, valor_aumento):
        if valor_aumento >= 1:
            percentual = valor_aumento / 100
        else:
            percentual = valor_aumento
        self.salario += self.salario * percentual

def converter_para_float(texto):
    texto_limpo = texto.replace('.', '').replace(',', '.')
    return float(texto_limpo)

n_func = input("Nome do Funcionário: ")
s_func_raw = input(f"Salário atual de {n_func} (ex: 1.612,00 ou 1612): ")
s_func = converter_para_float(s_func_raw)

c_func = input("Cargo: ")
func = Funcionario(n_func, s_func, c_func)

perc_raw = input(f"Porcentagem de aumento para {n_func} (ex: 6 ou 0,06): ").replace(',', '.')
perc = float(perc_raw)

func.aumentar_salario(perc)
print(f"Novo salário de {func.nome}: R$ {func.salario:.2f}")
```
